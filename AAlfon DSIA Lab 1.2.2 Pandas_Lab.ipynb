# 5. Visualization

#### 5.A Plot Histogram of Age 

````python
plt.hist(df.age , facecolor = 'green', alpha = 0.5)
````

#### 5.B BoxPlot of Age

````python
plt.boxplot(df.age)
plt.yscale('linear')
plt.grid(b=True, which='major', color='b', linestyle='-')
plt.grid(b=True, which='minor', color='r', linestyle='--')
plt.show()
````

````python
plt.scatter(df.age, df.length_of_service)
plt.show()
````
